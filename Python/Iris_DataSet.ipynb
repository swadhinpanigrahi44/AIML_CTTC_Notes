```python
#wiwe'll take the input only

ip = iris.drop('Flower', axis=1)
ip.head()
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

```python
#KMeans Clustering
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
km.fit(ip)
```

```python
km_pred = km.predict(ip)
km_pred
```

```python
centroid = km.cluster_centers_
centroid
```

```python
iris['Prediction'] = km_pred
iris
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length)
plt.show()
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.petal_length,c=km_pred)
plt.show()
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.petal_length, c=km_pred)
sns.scatterplot(x=centroid[:,0], y=centroid[:,2], s=200, color='k')
plt.show()
```

```python
sns.scatterplot(x=iris.sepal_width, y=iris.petal_width, c=km_pred)
sns.scatterplot(x=centroid[:,1], y=centroid[:,3], s=200, color='k')
plt.show()
```
