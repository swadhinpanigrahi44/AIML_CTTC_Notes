```python
Write a Python program to check whether the entered number is even or odd, as well as positive or negative
```

```python
n = int(input("Enter a number: "))

if n % 2 == 0:
    if n > 0:
        print("Even and Positive")
    elif n < 0:
        print("Even and Negative")
    else:
        print("Even and Zero")
else:
    if n > 0:
        print("Odd and Positive")
    elif n < 0:
        print("Odd and Negative")
    else:
        print("Odd and Zero")
```

# while loop

```python
syntax:
    initialize
        while condition:
            #body of while
            increment/decrement
```

```python
i=1
while i<=100:
    print(i)
    i +=1
```

```python
i=1
while i<=100:
    j=1
    factor=0
    while j<=1:
        if i%j==0:
            factor+=1
        j+=1
    if factor==2:
        print(i,end=',')
    i+=1

```

```python
i=1
while i<=100:
    
    j=1
    factor=0
    while j<=i:
        if i%j==0:
            factor+=1
        j+=1
    if factor==2:
        print(i,end = ',')
    i+=1
    
```

# break statement
the break is a keyword in pythin which is used to bring the programe control out of looop..

```python
for i in range(1,15):
    print(i,end=" ")
    break
```

```python
for i in range(1,15):
    if i%8==0:
        break
    print(i,end=" ")
```

# continue statement:the continue keyword return control of the iteration to the begining of the python for loop or python while looop.all remaining lines in the prevailling iteration of the loop are skipped by the continue keyword ,which returns execution to the beginning of the next iteration of the loop

```python
for i in range(1,15):
    print(i,end=" ")
    continue
```

```python
for i in range(1,15):
    if i%2==0:
        continue
    print(i,end=" ")
```
