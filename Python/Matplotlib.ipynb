```python
#importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#line plot
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.plot(x,y,color='c')
plt.title('line plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()
plt.show()
```

```python
x=[20,30,45,67,78,39]
y=[30,25,35,16,38,34]

plt.plot(x,y,'p-.',linewidth=2,markersize=20,color='b')
plt.title('line plot(??`?)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid()
plt.show()
```

```python
#bar plot
x=[20,30,45,67,78]
y=[30,25,35,16,38]

plt.bar(x,y,label='AIML',width=7,color='c')
plt.title('Bar Graph')
plt.xlabel('X-Axis',color='m')
plt.ylabel('Y-Axis',color='m')

plt.legend()
plt.show()
```

```python
x=[1,3,5,7,9]
y=[2,4,6,8,10]
plt.bar(x,y,label='bar',color='orange')

x=[1,3,5,7,9]


plt.plot(x,y,label='line')
plt.title('Bar& line Graph')
plt.xlabel('X-Axis',color='m')
plt.ylabel('Y-Axis',color='m')
plt.annotate("value",xy=(1,2.5),size=10,color='g') 
```

```python
#Histogram
ages=[20,19,13,12,16,18,22,23,29,40,35,36,45,48]
bins=[10,20,30,40,50]

plt.hist(ages,bins,histtype='bar',width=8)
plt.title('Histogram')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#scatter plot
x=[1,4,3,5,2,7,6,8,9]
y=[2,1,4,5,6,8,3,7,6]

plt.bar(x,y,color='g',width=0.1)
plt.scatter(x,y,s=200,color='red',marker='*',linewidth=6,edgecolors='red')
plt.scatter(x,y,s=20,color='gold',marker='o')
plt.title('scatter plot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
#Pie chart
slices=[8,10,4,3,5]
activities=['sleeping','phone','study','eat','play']
col=['gold','b','magenta','red','g']
plt.pie(slices,labels=activities,colors=col,startangle=90,explode=(0,0,0.3,0,0),shadow=True,autopct='%1.2f%%')

plt.title('Pie chart')
plt.show()
```
